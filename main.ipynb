```python
import pandas as pd

df = pd.read_csv("icpc-full.csv")

display(df.head(10))

print("Columnas del dataset:")
print(df.columns.tolist())

print("\nInformación del dataset:")
print(df.info())

print("\nEjemplo de universidades con medallas:")
print(df[['University', 'Country', 'Gold', 'Silver', 'Bronze']].head())

```

## Universidades con más medallas en ICPC World Finals

```python
import matplotlib.pyplot as plt

uni_medals = df.groupby('University')[['Gold','Silver','Bronze']].sum()

uni_medals['Total'] = uni_medals['Gold'] + uni_medals['Silver'] + uni_medals['Bronze']

top_unis = uni_medals.sort_values('Total', ascending=False).head(10)

top_unis[['Gold','Silver','Bronze']].plot(
    kind='bar',
    stacked=True,
    figsize=(12,6),
    color=['gold','silver','#cd7f32']
)

plt.title("Top 10 universidades con más medallas en ICPC World Finals")
plt.xlabel("Universidad")
plt.ylabel("Número de medallas")
plt.xticks(rotation=45, ha='right')
plt.legend(title="Tipo de medalla")
plt.tight_layout()
plt.show()
```

## Países con más campeones

```python
import matplotlib.pyplot as plt

winners_by_country = df[df['Rank'] == 1].groupby('Country').size().sort_values(ascending=True)

winners_by_country.plot(kind='barh', figsize=(10,6), color="skyblue")
plt.title("Países con más campeones ICPC")
plt.xlabel("Número de campeonatos")
plt.ylabel("País")
plt.show()
```

## Universidades ganadoras a través de los años

```python
champions = df[df['Rank'] == 1][['Year','University','Country']].sort_values('Year')

plt.figure(figsize=(12,6))
plt.scatter(champions['Year'], champions['University'], c='purple', s=100)
plt.title("Universidades campeonas a lo largo de los años")
plt.xlabel("Año")
plt.ylabel("Universidad")
plt.grid(True, linestyle='--', alpha=0.6)
plt.show()

display(champions)
```

## Universidades que más han repetido 

```python
repeated_champions = df[df['Rank'] == 1].groupby('University').size().sort_values(ascending=False).head(8)

repeated_champions.plot(
    kind='pie', 
    autopct='%1.1f%%', 
    figsize=(8,8), 
    startangle=90, 
    cmap="tab20"
)
plt.title("Universidades con más títulos ICPC")
plt.ylabel("")
plt.show()
```

## País con más oros, platas y bronces

```python
country_medals = df.groupby('Country')[['Gold','Silver','Bronze']].sum()
top_medals = country_medals.sort_values('Gold', ascending=False).head(10)

top_medals[['Gold','Silver','Bronze']].plot(
    kind='bar',
    figsize=(12,6)
)
plt.title("Top 10 países con más medallas (oro, plata, bronce)")
plt.xlabel("País")
plt.ylabel("Número de medallas")
plt.xticks(rotation=45, ha="right")
plt.show()

```

## Universidades más exitosas dentro de cada país

```python
import seaborn as sns
import matplotlib.pyplot as plt

uni_medals = df.groupby(['Country','University'])[['Gold','Silver','Bronze']].sum()
uni_medals['Total'] = uni_medals.sum(axis=1)

top_uni_per_country = uni_medals.reset_index().sort_values('Total', ascending=False).head(20)

heatmap_data = top_uni_per_country.pivot_table(
    index="University",
    columns="Country",
    values="Total",
    aggfunc="sum",
    fill_value=0
)

# Graficar
plt.figure(figsize=(12,6))
sns.heatmap(heatmap_data, annot=True, cmap="YlGnBu", fmt="g")
plt.title("Universidades más exitosas dentro de cada país")
plt.show()


```

## Ganadores por década

```python
df['Decade'] = (df['Year']//10)*10
decade_champs = df[df['Rank']==1].groupby(['Decade','Country']).size().unstack(fill_value=0)

decade_champs.plot(kind='bar', stacked=True, figsize=(12,6), colormap="tab20")
plt.title("Países campeones por década")
plt.xlabel("Década")
plt.ylabel("Número de campeonatos")
plt.show()
```

## Países debutantes por año

```python
first_appearance = df.groupby('Country')['Year'].min()
new_countries_per_year = first_appearance.value_counts().sort_index()

new_countries_per_year.plot(kind='bar', figsize=(12,6), color="teal")
plt.title("Número de países que debutaron cada año en el ICPC")
plt.xlabel("Año")
plt.ylabel("Número de nuevos países")
plt.show()
```

## País con más universidades ganadoras

```python
import matplotlib.pyplot as plt

uni_gold = df[df['Gold'] > 0].groupby(['Country','University']).size().reset_index(name='Wins')

country_uni_counts = uni_gold.groupby('Country')['University'].nunique().sort_values(ascending=False).head(8)
top_countries = country_uni_counts.index

top_uni_gold = uni_gold[uni_gold['Country'].isin(top_countries)]

top_uni_gold = top_uni_gold.sort_values(['Country','Wins'], ascending=[True,False])

plt.figure(figsize=(12,8))

for i, country in enumerate(top_countries):
    subset = top_uni_gold[top_uni_gold['Country'] == country]
    plt.barh(subset['University'] + " (" + subset['Country'] + ")", subset['Wins'], label=country)

plt.title("Universidades ganadoras de oro en los países con más universidades ganadoras")
plt.xlabel("Número de oros")
plt.ylabel("Universidades (con país)")
plt.legend(title="País", bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()

```

## Ranking de la Universidad de La Habana

```python
import pandas as pd
import matplotlib.pyplot as plt

name_variants = [
    "Universidad de La Habana",
    "University of Havana",
    "Universidad de la Habana", 
    "UH" 
]

mask = False
for nv in name_variants:
    mask = mask | df['University'].str.contains(nv, case=False, na=False)

uh_df = df[mask].copy()

if uh_df.empty:
    print("No se encontraron registros para la Universidad de La Habana. Revisa el nombre exacto en df['University'].unique()")
else:

    uh_yearly = (
        uh_df.groupby('Year', as_index=False)['Rank']
             .min()
             .sort_values('Year')
    )

    display(uh_yearly.rename(columns={'Rank': 'Mejor_Rank'}))

    plt.figure(figsize=(10,5))
    plt.plot(uh_yearly['Year'], uh_yearly['Rank'], marker='o')
    plt.gca().invert_yaxis()
    plt.title("Universidad de La Habana – Mejor lugar por año en ICPC World Finals")
    plt.xlabel("Año")
    plt.ylabel("Lugar (Rank)")
    plt.grid(True)
    plt.tight_layout()
    plt.show()
```
